```python
#BRT function to get lambda's
def BRT(alpha, c, X, Y):
    sigma = np.std(Y)
    (n,p) = X.shape
    lamb = 2*c*sigma*norm.ppf(1-(alpha)/(2*p))*1/np.sqrt(n)
    return lamb
#BCCH function to get lambda's
#def BCCH(alpha, c, X, Y):
#    (n,p) = X.shape
#    scale = np.max(np.mean(np.dot(np.dot(X,X.T),np.dot(Y,Y.T)), axis = 0)) **0.5
#    lamb_temp = 2*c*norm.ppf(1-(alpha)/(2*p))*scale/np.sqrt(n)
#    lasso_temp = Lasso(alpha=lamb_temp, max_iter= 1000000, tol=0.1)
#    lasso_temp.fit(X ,Y)
#    prediction = lasso_temp.predict(X)
#    residual = (Y - prediction)
#    scale2 = np.max(np.mean(np.dot(np.dot(X,X.T),np.dot(residual,residual.T)), axis = 0)) **0.5
#    lamb = 2*c*norm.ppf(1-(alpha)/(2*p))*scale2/np.sqrt(n)
#    return lamb

def BCCH(X_tilde, Yt):
    alpha = .05
    c = 1.1
    n = X_tilde.shape[0]
    p = X_tilde.shape[1]

    llambda_pilot = ( (2 * c) / np.sqrt(n) ) * norm.ppf(1 - (alpha/(2*p))) * np.sqrt(max(np.sum(X_tilde**2 * Yt.reshape(-1,1)**2, axis = 0))/n)

    lasso_pilot = Lasso(alpha=llambda_pilot, max_iter = 10000)
    lasso_pilot.fit(X_tilde, Yt)
                    
    epsilon = Yt - lasso_pilot.predict(X_tilde).reshape(-1,1)

    llambda = ( (2 * c) / np.sqrt(n) ) * norm.ppf(1 - (alpha/(2*p))) * np.sqrt(max(np.sum(X_tilde**2 * epsilon.reshape(-1,1)**2, axis = 0))/n)

    return(llambda)

```

# Double Lasso No interaction terms-----------------50 independent Vars

```python
X = data.copy().drop(['DEPENDENT GROWTH RATE', 'Entity'], axis=1)

D = X['Apr_Temp']
Z = X.drop(['Apr_Temp'], axis=1)

Y = data['DEPENDENT GROWTH RATE']
#Y = np.array(Y).reshape(-1, 1)
```

```python
def processing(df):
    df = pd.get_dummies(df, columns=['Region'])
    scaler = preprocessing.StandardScaler()
    df =  scaler.fit_transform(df)
    return(df)
```

```python
X = processing(X)
Z = processing(Z)

scaler = preprocessing.StandardScaler()
D =  scaler.fit_transform(np.array(D).reshape(-1, 1))
```

# First-Stage

```python
alpha = .05
c = 1.1
```

```python
Z.shape
```

```python
lamb1st = BRT(alpha = alpha, c = c, X=Z, Y=D)
#lamb1st = BCCH(Z,D)
lasso1st = Lasso(alpha=lamb1st, max_iter = 1000000, tol=0.1)
lasso1st.fit(Z,D)
```

```python
psi = lasso1st.coef_.reshape(-1, 1)
```

```python
print('First Stage: lambda = ',lamb1st , "" )
```

# Second-Stage

```python
alpha = .05
c = 1.1
lamb2nd = BRT(alpha = alpha, c = c, X=X, Y=Y)
#lamb2nd = BCCH(X ,np.array(Y).reshape(-1,1))
lasso2nd = Lasso(alpha=lamb2nd, max_iter = 1000000)
lasso2nd.fit(X,Y)
```

```python
gamma = lasso2nd.coef_.reshape(-1, 1)[1:,:]
alpha = lasso2nd.coef_.reshape(-1, 1)[0,:]
```

```python
print('Second Stage: lambda = ',lamb2nd , "" )
```

```python
print("Only ", len(psi[psi==0]), " of the ", X.shape[1], " variables were sent to 0.")
```

### Because these llambdas are so small, none of the coefficients are sent to zero. Which implies no variable selection occurs and Post-Lasso is equivalent to OLS.

# Alpha Tilda Calculation

```python
Y = np.array(Y).reshape(-1, 1)
```

```python
numerator =  np.sum( ( Y - np.dot(Z, gamma) ) * ( D - np.dot(Z, psi) ) )
denominator = np.sum(D * D - np.dot(Z,psi))

alphatilda = numerator/denominator
```

```python
print( "The estimated alpha using Lasso is", round(alpha[0],2), "When using Double Lasso it is",round(alphatilda,2))
```

```python
epsilon = Y - lasso2nd.predict(X)
nu = D - np.dot(Z,psi)
```

```python
V = np.mean(epsilon**2 * nu**2) / np.mean(nu**2)**2
SE = np.sqrt(V) / np.sqrt(len(Y))
```

```python
print("The 95% C.I. for alpha-tilda is : ", "[", alphatilda - 2 * SE , ",", alphatilda + 2 * SE,"]")
```

# Compare to OLS 50 Vars

```python
import statsmodels.api as sm
ols = sm.OLS(Y, X)
ols_result = ols.fit()
# Now you have at your disposition several error estimates, e.g.
olsSE = ols_result.HC0_se[0]
olsalpha = ols_result.params[0]
print("The 95% C.I. for ols-alpha is : ", "[", olsalpha - 2 * olsSE , ",", olsalpha + 2 * olsSE,"]")
```

# Double lasso with Interaction terms----------1325 Dependent Variables

```python
X = data.copy().drop(['DEPENDENT GROWTH RATE', 'Entity'], axis=1)
Y = data['DEPENDENT GROWTH RATE']

def processing(df):
    df = pd.get_dummies(df, columns=['Region'])
    poly = PolynomialFeatures(2, include_bias = False)
    poly_nparray = poly.fit_transform(df)
    scaler = preprocessing.StandardScaler()
    df =  scaler.fit_transform(poly_nparray)
    return(df)

X = processing(X)
Z = X[:,1:]
D = X[:,0].reshape(-1,1)

alpha = 3
c = 1.1

lamb1st = BCCH(Z,D)
#lamb1st = BRT(alpha = alpha, c = c, X=Z, Y=D)
lasso1st = Lasso(alpha=lamb1st, max_iter = 1000000, tol=0.1)
lasso1st.fit(Z,D)

psi = lasso1st.coef_.reshape(-1, 1)
print('First Stage: lambda = ',lamb1st , "" )

alpha = .05
c = 1.1

#lamb2nd = BRT(alpha = alpha, c = c, X=X, Y=Y)
lamb2nd = BCCH(X ,np.array(Y).reshape(-1,1))
lasso2nd = Lasso(alpha=lamb2nd, max_iter = 1000000)
lasso2nd.fit(X,Y)

gamma = lasso2nd.coef_.reshape(-1, 1)[1:,:]
alpha = lasso2nd.coef_.reshape(-1, 1)[0,:]
print('Second Stage: lambda = ',lamb2nd , "" )

Y = np.array(Y).reshape(-1, 1)
numerator =  np.sum( ( Y - np.dot(Z, gamma) ) * ( D - np.dot(Z, psi) ) )
denominator = np.sum(D * D - np.dot(Z,psi))
alphatilda = numerator/denominator

print( "The estimated alpha using Lasso is", round(alpha[0],2), "When using Double Lasso it is",round(alphatilda,2))

epsilon = Y - lasso2nd.predict(X)
nu = D - np.dot(Z,psi)

V = np.mean(epsilon**2 * nu**2) / np.mean(nu**2)**2
SE = np.sqrt(V) / np.sqrt(len(Y))

print("The 95% C.I. for alpha-tilda is : ", "[", alphatilda - 2 * SE , ",", alphatilda + 2 * SE,"]")

print("Only ", len(psi[psi==0]), " of the ", X.shape[1], " variables were sent to 0.")
```
